Make functions for calculating the two data matrixes $F$ & $Y$

```python
def fY(data,all_ = 0):
    # Do a type check
    if type(data) == pd.core.frame.DataFrame:
        data = data.values
    elif type(data) != np.ndarray:
        raise TypeError("Expected type 'numpy.ndarray' or 'pandas.core.frame.DataFrame' was given '{}'".format(type(data)))
    ts = data[:,1:].T #Species abudance Y
    grow = np.ones([1,ts.shape[1]])# Growth rate Y
    Y_all = np.append(grow,ts,axis=0) # has L+1 time elements
    if all_:
        return Y_all # return all L+1 elements
    else:
        return np.delete(Y_all, -1, axis = 1) # return L elements (removing the last one)

def fF(data):
    if type(data) == pd.core.frame.DataFrame:
        data = data.values
    elif type(data) != np.ndarray:
        raise TypeError("Expected type 'numpy.ndarray' or 'pandas.core.frame.DataFrame' was given '{}'".format(type(data)))
    dt = np.diff(data[:,0])
    ts = fY(data,1)[1:,:]# Have all L+1 elements inorder to caluclate the L long F
    dln = np.diff(np.log(ts))
    return dln/dt
```

```python
Y=fY(Timeseries['timeseries'])
F=fF(Timeseries['timeseries'])
lambda_mu = 0.01
lambda_M = 0.01
D=np.diag( np.append( lambda_mu, np.repeat(lambda_M,F.shape[0]) ))
StartTheta=np.random.normal(1,0.2,[F.shape[0],Y.shape[0]])
```

```python
def RidgeGradientCheck(Theta):
    if ('Y' or 'F' or 'D') not in globals():
        raise SystemError("Data matrixes 'Y','F' and 'D' need to be defined globaly")
    Check=np.array([Y.shape[1]!=F.shape[1], Theta.shape[0]!=F.shape[0], Theta.shape[1]!=Y.shape[0], Theta.shape[1]!=D.shape[0]])
    if any(Check):
        vs=["Y vs F", "Theta vs F", "Theta vs Y", "Theta vs D"]
        raise ValueError("Dimensions for: {}, do not match. Cannot perform matrix calculation".format(vs[Check]))
    return 2*(np.dot( np.dot(Theta,Y)-F ,Y.T) + np.dot(Theta,D))

def RidgeGrad(Theta):
    return 2*(np.dot( np.dot(Theta,Y)-F ,Y.T) + np.dot(Theta,D))
```

```python
def Solution():
    return np.dot(F,np.dot(Y.T, np.linalg.inv(np.dot(Y,Y.T) - D) ))
```

```python
def total_error(Theta1,Theta2):
    error=np.abs(Theta1-Theta2)
    return(np.sum(error))
    
```

```python
alpha_exp=np.arange(-4.80,-4.784,.001)
alpha_list= np.power(10,alpha_exp) #0.1 to include 0 
Sol=Solution() #Solution matrix
Errors=np.array([])
for a in alpha_list:
    Guess = DescSpartan(Start=StartTheta, gradientmodel=RidgeGrad, alpha=a, max_itt=1e2,verbose = 0)
    Errors = np.append(Errors,total_error(Sol,Guess))
    
fig, ax = plt.subplots()
ax.plot(alpha_exp, Errors)
ax.set(xlabel='log10(alpha)', ylabel='Total error Theta',
       title='Alpha dependency')
ax.grid()
plt.show()
print(alpha_exp)
```

There seems to be a sweet spot at -4.785 Where the erros is minimum. (when max_itt=100). After that It blows up. 

However if let the max_itteration go higher. The ideal alpha is higher. 
For max_itt=1e3: we have an ideal alpha= -4.781(-4.780) technically from -4.780 it starts blowing up but very slowly in the beginning. 

So lets safely say exponent -4.785 is the highest alpha we can have where the things does not blow up.

# Effects of noise and lambda

So in the previous section we evaluated how the gradient descent performed, and how it converged to Solution(). We came to the conclusion that the accuracy was ok-ish. But could be better, and this maybe has to be looked into further

But now we come to the more important part. How the minimum deviates from the real interaction matrix! There is no use in the best minimizing method if the minimum its self is far away from the real. 

Now this ofcourse has to do with the noise that is added in our simulation. Because the noise lets the curve go in directions that the gLV can never capture. The gLV interaction matrix that is really captured is one that spirals downwards to steady state. 

So, depending on where we are in the phase space, from a given time entry our derived model predicts that the next should go down. But then suddenly noise kicks in and the next entry is up... And this happens all the time. So its logical to indeed see that the theoretical minimum can never be correct, since the data we have could never be captured by a gLV. And so the outcome fit, also fits the noise. which is wrong.

The fitted model might even not be stable... 

```python
print(glv_minimal_code.is_stable(Timeseries['interaction_matrix'],0,0))
print(glv_minimal_code.is_stable(Theta_5[:,1:],0,0))
```

But it is in this case...

### noise is baked in
The other thing that makes this difficult is that noise is baked in. What I mean by this is that the model is not simply just simulated and then we add a noise filter on top of it like:

$$Y=f(X)+\epsilon$$

This is not the case. Because here the noise is centerd around the ideal curve described by $Y=f(X)$. And this is also where most statics book focus around on how to deal with the "intrinsic" variablity, or also often called *irreducable* variablity. 

But here it is different. The noise effects the curve.... Since every point is seen as a new starting entry. So if noise kicks our prediction wide of. Well this now unusual point is our new point and the data is simulated from that point onward. In the latter case this high noise point would kick out one individual point, but our curve is still were it needs to be and thus the next point would there follow more closely the original curve. 



The thing that noise is baked in it, is both a blessing and a curse. 

**curse:**
<br>
It is a curse, in the sense that it is now way harder to fit the data to it. Since most fitting methods assume noise of the first kind, and are developed around it. Like the gradient descent. The second aspect to the curse is that we can hardly make any prediction. 

The most straight forward effect of this actually already show above in the first figure. Here the gLV actually predicts that it goes very quickly to the steady state. Then the prediction of the gLV is very boring... just steady state and constant. While clearly the effects of noise above kick it constantly out of equilibrium. It is then because of the gLV it ventures back to the steady state. 

So inreality there is this constant battle going on between gLV and noise, while our prediction is this dead heartbeat. So our prediction is always no where close... because the real curve is some where far out. Hence our prediction makes no sense, and it also **makes no sense to calculate the Mean Square Error (MSE)**. Since the model is not just a mear noise filter added on top of an ideal curve. 

**Blessing:**
<br>
Even though I sounded very cynical about the whole situation. There are some upside to it. Because of this constant battle between noise and gLV. We see the gLV constantly in action, as opposed to our dead heartbeat which bears no information at all about the gLV. 

So because of this we can start extracting information about the gLV while the "ideal curve" would have not shown us anything. + the real curve is also not realistic, it is the noise model that we have that is realistic. 

How might one extract information of the gLV from the noisy curve? Say, that our species abundance happend to be at the steady state of the gLV model. And we get hit by a big noise peak and are send off. We are now situated far away from the stable state and the gLV model dictates the system to spiral back towards it. Now imagine for a breef moment that the noise is polite and is not a large. The subsequent points will follow closely the curve predicted by the gLV starting from the largely kicked off point. These points could be used to infer the gLV. 

Ofcourse we do not know the proportions in change of the state duo gLV and noise. But one can actually try to make educated geusses. If a point is suddenly largely off what the gLV predicts -> then we can assume that this was caused by noise and we could maybe penalise this point by adding a low weight to it in the fitting. (which is not done here at all since gradient descent is not designed for this fitting). 

If the noise had the form of turning of and on again. It would be equivalent to having no noise. Since we could then just pick out the parts where there is no noise. And thus have many situation in which we could see the gLV work with no error. 
Ofcourse this is not the case and so it more finding out when the noise is small and thus to determine when could use the data and when not.

## investigation minimum vs real on different noise levels

```python
# difference of the current minimum and the real normalised by the matrix element
plotM(abs((Sol-real)/abs(real)),"Distance minimum from real normalised by real")
# Total error
print("Total error = {}".format(total_error(Sol,real)))
print(real)
```

We see very high distance from the real interaction matrix. most of the same order as the matrix element self. But this means that it is not good... Also it seems to have problems with the small offdiagonal elements. First column and diagonal are pretty dark. 

```python
noise_levels = np.arange(0,0.055,0.005)
#noise_levels = np.array([0])

def SolutionFYD(F,Y,D):
    return np.dot(F,np.dot(Y.T, np.linalg.inv(np.dot(Y,Y.T) - D) ))

Errors=np.array([])
for noise in noise_levels:
    ts = glv_minimal_code.main(seed=1,noise=noise,sigma=0.05)
    theta_real = np.append(ts['growthrate'],ts['interaction_matrix'],axis=1)
    lambda_mu = 0
    lambda_M = 0
    D=np.diag( np.append( lambda_mu, np.repeat(lambda_M,F.shape[0]) ))
    theta_sol = SolutionFYD(fF(ts['timeseries']),fY(ts['timeseries']),D)
    Errors = np.append(Errors,total_error(theta_real,theta_sol))
    
fig, ax = plt.subplots()
ax.plot(noise_levels, Errors)
ax.set(xlabel='noise', ylabel='Total error',
       title='Effects noise')
ax.grid()
plt.show()
```

```python
plotM(abs(theta_real-theta_sol)/abs(theta_real),"test")
np.round(real,4)
```
