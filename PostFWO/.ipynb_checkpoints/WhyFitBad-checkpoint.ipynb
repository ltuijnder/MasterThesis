Define the a functiont that can plot the 3D data.

```python
def plot3DY(Y,X,Exp = 0,Ynumber = 0):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    
    YData = Y[Exp,:,Ynumber]
    testX = X[Exp,:,1]# Remove the zeroth column since it are just ones. 
    
    ax.plot(testX,YData,"*")
    
plot3DY(fitObj.Y,fitObj.X)
```
