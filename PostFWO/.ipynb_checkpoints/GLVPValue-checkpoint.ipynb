# Generating the data

First, we need a way of creating data. As has been discussed extensivly, we are going to simulate the GLV with steady state at [1,1,1, 1] and we will NOT centere the data and accept the parameters as growth rate and self interaction and interaction from other species !

This data generation is done in a script named _generateGLV_ so import it.

```python
from generateGLV import *
```

The timeseries itself is an object of the class _TS_GLV_ This class contains all the needed information of the timeseries. Like the parameters, the model,noise contribution. All the different settings. One can even reset the object and generate again.

However, it is meant that the settings of the timeseries are provided at the creation of the object and the settings should be de then left unchanged. Such that a given object is associated with a given configuration. 

Note that appart from just the amount of species. We also specify the number of experiments. This is basically inorder to avoid generating multiple objects having all the same configuration settings. We just create one object with which this settings is associated and number of experiments then tells us how often we should generate a timeseries with that same specific setting. 

```python
# Settings Generation

noiseSetting = {
    "noiseType" : "LangevinLinear",
    "noiseStrength" : 0.02
}

pertuSetting = {
    "period" : 50,
    "strenght" : 0.05
}

genSetting = {
    "interactionStrenght" : 1
}

numberSpecies = 5
numberExperiments = 3

# Generate

TS = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting,pertuSetting)
TS.generate()
TS.plot(1)# Plot the zeroth experiment.
```

To extract the result simply acces the .result value of the TS object.

### Observation made regarding the square fitted noise

```python
np.sqrt(fitObj.varEst)
```

When the timestep = 0.01 and noise = 0.01 -> Fitted std = 0.1
But when timestep =0.0001 then fitted std = 1. Thus a factor 100 decrease in timestep is a factor 10 increase in the noise.

-> 100 to 10 comes from the square root ok. But why does the level of std increase? Is this logical? 

Aah I think I realise why. Thus the timesteps are now 100 smaller. However, the noise only scales down with 10 (because of the sqrt(timestep) -> But this means that the noise is effect 10 times bigger per timestep.

This also explains why it before first fitted a noise of 0.1 because we were working with timesteps 0.01.

So model contribution goes down by a factor 100.
But but noise only goes down by 10... 
-> This is not correct. -> Because now the noise becomes basically just stronger with a smaller timestep. With very small timesteps then the only contribution = noise.
-> And the timeseries should not depened on the choosen timestep. 

But this is not correct :( Because we wrote our GLV as dlog/dt = b0 + sum(X * b1) + epsilon. with epsilon a normal distribution equal to our strenght. And now the strenghts are more then what it is supposed to be. 

-> However by taking square root it seems however that the model indeed stays independed of the timestep. When I artificially change an order extra with the noise strenght (what whould have happend if the timeseries changed linearly and noise stayed the same) the qualitif results changes significant. So the sqrt seems to be correct -> But this does not take away the fact that the noise contribution goes up linearly!

## Try and debug the bad inference with manually generated data:
import the correct model

```python
from GeneralDataGen import *
```

Create the object

```python
numspecies = 5
testObject = Data(numspecies,numspecies)
```

Let set the noise level

```python
noiseValue = 0.05
testObject.setNoiseStd(np.repeat(noiseValue,testObject.outputDim))
```

Generate a costum interaction matrix

```python
intercept = np.ones(numspecies).reshape(1,numspecies)
interaction = np.full((numspecies,numspecies),0.5)
# Add a few custom ones to convince our self
interaction[2,4]=noiseValue*0.5 # One clearly below the noise level
interaction[2,3]=10 # And one clearly large
np.fill_diagonal(interaction,-1)
parameter = np.append(intercept,interaction,axis=0)
testObject.setModelParameters(parameter)
testObject.coefMatrix
```

```python
testObject.numberOfObservations = 500 
testObject.generate()
#testObject.plotData(0,4) # plotting does not really help when there are a lot of observations
```

-----
Now let the coeff matrix be fitted by the Fit class by setting the data type explicitly to "Data"

```python
fitDataObj = fitGLV(testObject,typeInput="Data")
```

Now evaluate if the fit was decent.

```python
fitDataObj.fitLinear()
```

```python
expNumb = 1
print(f"Fitted coefMatrix of exp.{expNumb}:")
print(np.round(fitDataObj.BEst[expNumb],4))
```

It looks like the fit is perfect! Next we evaluate if the variation of the Beta estimated

```python
fitDataObj.computeVarBEst()
plotM(fitDataObj.varBEst[expNumb],"")
```

The values look plausable so I am willing to take on it that this is correct.

Next we make the null hypthesis. This should all be significant non zero. 

```python
fitDataObj.computeNullHypo()
fitDataObj.plotNullHypo(expNumb)
fitDataObj.pNull[expNumb,3,4]
```

This looks correct. But now the big question !! 

Does the hypothesis with it self look correct?

```python
out = fitDataObj.hypo(testObject.coefMatrix,ExpNum=expNumb)
out[1][3,4]
```

Great! The when doing a hypothesis test with the correct matrix it produces the wanted result. Namely, that no parameter is significant and hence we cannot reject the null hypothesis. But this is also correct since the null hypothesis in this case si the correct one! 

It had me worried in the beginning because. Looking at the variation values above. They seemed VERY small, 0.00007 -> $\sigma$ = 0.0083. But actually this is still inline with what we see with the fitted values. -> Largest deviations are 0.016. Which is just twice the sigma. Hence the fits are correct!

Also, this text was written when there we did not incorporate the look else where effect. We then indeed saw the correct behaviour of that most data is <2 sigme (95%), but a few where in the [2,3] sigma region. Which is expected when you have 30 parameters. You are bound the eventually have 1/20 chance.

### Some further observations that are made:

_Costum parameters:_

When I change the one parameter really high and another parameter really low to the level of ( $0.5*noise$ ). Then we indeed see indeed that the it becomes harder for the fit to find that small parameter. We now get a 3.5 sigma chance (without look elsewhere) for the null hypothesis. -> When we would incorporate look else where it can hardly be distanced vieuwed seperatly.

When doing then the hypothesis with the real matrix. Nothing changed (like litterly the value does not change)

_High number of Experiments:_

To mimic like what we fit with the TimeSeries that we have a huge number of observations. Thus we also now made a fit with 10 000 observations.

Result: The result is really good! With 10 000, the result became just more clear! It became more confident of its results. The fluke of 3.5 $\sigma$ with the small parameter in the NullHypo -> went away. 

And now in the the real hypothesis. All parameters are <2 sigma.
What is still strange is that it starts to fit small parameter to 0.0222 while the real one is 0.025. Diff = 0.025-0.022 = 0.003

While the variance is 0.0000032 -> $\sigma$ = 0.0017. So what seemed to be a wrong fit first is still within confidence!

_Incorperation look else where effect:_

When incorporating the look else where effect -> Fluctuations in real hypothesis go away and everything stays nicely under the 2 sigma. There would only be a real life chance of 1/20 that we would see a higher value. Meaning if we would do 20 experipents. We expect that on average 1 of those would contain a parameter whose p-value is higher then 2 sigma. But, then again this just purely happend by chance.

The not so nice thing about look else where effect though, is that we can now have less confidence in the null hypothesis to reject stuff. Namely the sigma is now went from 3.5 $\sigma$ -> 2.4 $\sigma$ . But now we should also be more serious about our result. 2.4 may not sound like much... But it still means that there was only a probability of 2.4 $\sigma$ = 1.3 % . That this happend by chance... Which is very unlikely but still the hard threshold should be 3 sigma. But I do not know if this is really handy with lows amount of data... this look else where effect.

When looking at the second experiment: We again have a value of 1.4 % now. Hence, for two completly independed experiments we had a +- 1.5 % Which would actually mean the chance of this happing by random is $0.015*0.014=0.02\%$ . Which is far above the 3 $\sigma$ . Hence we can reject that parameter. But ofcourse in real live you do not have two experiments.
