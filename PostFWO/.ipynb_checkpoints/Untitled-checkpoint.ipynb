# Generating the data

First, we need a way of creating data. As has been discussed extensivly, we are going to simulate the GLV with steady state at [1,1,1, 1] and we will NOT centere the data and accept the parameters as growth rate and self interaction and interaction from other species !

This data generation is done in a script named _generateGLV_ so import it.

```python
from generateGLV import *
```

The timeseries itself is an object of the class _TS_GLV_ This class contains all the needed information of the timeseries. Like the parameters, the model,noise contribution. All the different settings. One can even reset the object and generate again.

However, it is meant that the settings of the timeseries are provided at the creation of the object and the settings should be de then left unchanged. Such that a given object is associated with a given configuration. 

Note that appart from just the amount of species. We also specify the number of experiments. This is basically inorder to avoid generating multiple objects having all the same configuration settings. We just create one object with which this settings is associated and number of experiments then tells us how often we should generate a timeseries with that same specific setting. 

```python
noiseSetting = {
    "noiseType" : "LangevinLinear",
    "noiseStrength" : 0.02
}

pertuSetting = {
    "period" : 20,
    "strenght" : 0.1
}

genSetting = {
    "interactionStrenght" : 0.02
}

numberSpecies = 5
numberExperiments = 3

TS = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting,pertuSetting)
TS.generate()
TS.plot(0)# Plot the zeroth experiment.
```

To extract the result simply acces the .result value of the TS object.

```python
TimeseriesExp1 = TS.result[1] # acces experiment 1
print(TimeseriesExp1.shape)
```
