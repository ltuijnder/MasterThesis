```python
def F1(currentState,dt,F):
    return dt*F(currentState)

def F2(currentState,dt,F):
    currentState_F1=currentState+F1(currentState,dt,F)/2
    return dt*F(currentState_F1)

def F3(currentState,dt,F):
    currentState_F2=currentState+F2(currentState,dt,F)/2
    return dt*F(currentState_F2)

def F4(currentState,dt,F):
    currentState_F3=currentState+F3(currentState,dt,F)
    return dt*F(currentState_F3)

def RK4_diff(currentState,dt,F):
    return F1(currentState,dt,F)/6+2/6*(F2(currentState,dt,F)+F3(currentState,dt,F))+F4(currentState,dt,F)/6
```

Test it with harmonic oscilation:

```python
# F of the harmonic oscilation:
def harmonicOscillator(currentState,period):
    # currentState y = [x,v]
    # => dy/dt = F = [v,-period^2*x]
    return np.array([currentState[1], -1 * period**2 * currentState[0]])

period = 0.5
initialState = np.array([1,0])

Fharmonic = lambda state: harmonicOscillator(state,period)
```

```python
# euler:
harmonicTS = Timeseries(F=Fharmonic, initialStates=initialState,
                        timestep= 0.1, noiseStrength=0.0, integrationType = "Euler")
harmonicTS.generate()
harmonicTS.plot()
```

```python
# RK4:
# euler:
harmonicTS = Timeseries(F=Fharmonic, initialStates=initialState,
                        timestep= 0.1, noiseStrength=0.0, integrationType = "RK4")
harmonicTS.generate()
harmonicTS.plot()
```
