```python
# Verification 1.
noiseSetting = {
    "noiseType" : "LangevinLinear",
    "noiseStrength" : 0.000
}
pertuSetting = {
    "period" : 1000,
    "strenght" : 0.05
}
genSetting = {
    "interactionStrenght" : 0.00}

numberSpecies = 5
numberExperiments = 1

# Generate
TS_NoNoiseNoInter = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting, pertuSetting)
TS_NoNoiseNoInter.generate()


genSetting = {"interactionStrenght" : 1}
TS_NoNoiseInter = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting, pertuSetting)
TS_NoNoiseInter.generate()

genSetting = {"interactionStrenght" : 0.0}
noiseSetting = {
    "noiseType" : "LangevinLinear",
    "noiseStrength" : 0.01
}
TS_NoiseNoInter = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting, pertuSetting)
TS_NoiseNoInter.generate()

genSetting = {"interactionStrenght" : 1}

TS_NoiseInter = TS_GLV(numberSpecies,numberExperiments,noiseSetting,genSetting, pertuSetting)
TS_NoiseInter.generate()

time = np.arange(0,100,0.01)
```

```python
expoNoise = np.arange(-5,-0,0.5)
noises = np.power(10,expoNoise)
expoInteraction = np.arange(-4.5,0.5,0.5)
interactions = np.power(10,expoInteraction)
pertubations = np.array([10000,20,2])
subSampleSteps = np.array([1,2,3,4,5,7,9,11,13,15,20,25,30,35,40,60,80,100,120,140])

fig, axs = plt.subplots(nrows=2, ncols=3, sharex=True, sharey=True, figsize=(15,10))
fig.suptitle(f"Summary performance for different parameter types", fontsize=22)

ParTypeName = ["interaction","growth rate","self interaction"]
ParType = ["I","G","S"]

VariableName = ["Percentage significant","Median significance"]
Variable = ["percent","median"]

stepSamples = (1,1,1)
Pertubation = (0,0)

pathToStorage = "DataStorage/FixedCode/"

for row, axs_row  in enumerate(axs):
        for column, ax in enumerate(axs_row):
            # create labels
            if row==0:# The top row
                ax.set_title(f"Parameter type = {ParTypeName[column]}",fontsize=15)
            if column==0: # The right side
                ax.set_ylabel(f"",fontsize=15)
            
            l = np.where(subSampleSteps==stepSamples[column])[0][0] # get back the correct index
            
            medianData = np.load(pathToStorage+"MEDIAN"+Variable[row]+ParType[column]+".npy")
            
            # Plot the data
            if Variable[row]=="percent":
                imPercent = ax.matshow(medianData[Pertubation[row],:,:,l],cmap="jet",vmin=0, vmax=1)
            else:
                imMedian = ax.matshow(medianData[Pertubation[row],:,:,l],cmap="jet",vmin=0, vmax=6)
# Tick lables for some weird reason have to be done this way :/ 
rightSideAxes = (axs[0][0],axs[1][0])
lowerSideAxes = (axs[1][0],axs[1][1],axs[1][2])
topSideAxes = (axs[0][0],axs[0][1],axs[0][2])
for ax in rightSideAxes:
    ax.set_yticks(np.arange(len(expoNoise)))
    ax.set_yticklabels(labels=expoNoise,fontsize=12)
for ax in lowerSideAxes:
    ax.set_xticks(np.arange(len(expoInteraction)))
    ax.set_xticklabels(labels=expoInteraction,fontsize=12)
    ax.tick_params(axis="x", bottom=True, top=True, labelbottom=True, labeltop=False)
for ax in topSideAxes:
    pass
    #ax.set_xticks(np.arange(7))
    #ax.set_xticklabels(labels=interStength)
    #ax.tick_params(axis="x", bottom=True, top=True, labelbottom=False, labeltop=True)
# Add color bar
cax = fig.add_axes([0.9, 0.515, 0.03, 0.35])
cbar = fig.colorbar(imPercent, cax=cax)
cbar.set_label("Fraction significant",fontsize=15)

cax = fig.add_axes([0.9, 0.135, 0.03, 0.35])
cbar = fig.colorbar(imMedian, cax=cax)
cbar.set_label(r"Median significance ($\sigma$)",fontsize=15)


# Add x and Y label
fig.text(0.5,0.07, "Log. Interaction strength", ha="center", va="center",fontsize=17)
fig.text(0.09,0.5, "Log. noise strength", ha="center", va="center", rotation=90,fontsize=17)  
# adjust space between pltos
    
plt.subplots_adjust( wspace=0, hspace=0.02)
plt.savefig("ThesisFigures/SimulationData/SummaryPerformance.png")
```
